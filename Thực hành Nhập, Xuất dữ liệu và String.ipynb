### Bài 3
Nhập vào 'tien_luong_theo_gio' va 'so_gio' trong 1 tháng.và 'tien_thuong'. sau đó in ra tổng thu nhập trong tháng biết rằng tổng thu nhập = tiền lương theo giờ * số giờ + tiền thưởng

```python
hour_wage = float(input("Please input your hourly wage:"))
working_hour = float(input("Please input your working hour:"))
incentive = float(input("Please input your incentive:"))
Total_monthly_wage = hour_wage * working_hour + incentive
print(f'{Total_monthly_wage}USD')
```

### Bài 4
Nhập vào một string và một số x. hãy xóa x ký tư đầu tiên của string,
Ví dụ string = "Hello Python", x = 3
output 'lo Python'

```python
string = input("Please type here: ")
x = int(input("Số ký tự bị xóa: "))
string = string[x:]
print(f"String sau khi cắt là: {string}")
```

### Bài 5
Nhập vào một string. kiểm tra xem đây có phải là string đối xứng hay không

Note: Chuối đối xứng là chuỗi có thứ tự giống nhau từ trái sang phải và ngược lại
ví dụ: abba, abcddcba

```python
string = input("Nhập string: ")
reverse_string = string[::-1]
print(string == reverse_string)
```
