### Thực hành vẽ hình chữ nhật

```python
import turtle
t = turtle.Turtle()

m = 100
n = 150

t.forward(m)
t.left(90)
t.forward(n)
t.left(90)
t.forward(m)
t.left(90)
t.forward(n)
t.left(90)
```

### Thực hành vẽ ngôi nhà mơ ước

```python
import turtle 
pen = turtle.Turtle()
pen.speed(10)

# vẽ mặt trời
pen.up()
pen.goto(60,150)
pen.down()
pen.pensize(3)
pen.pencolor("black")
pen.fillcolor("yellow")
pen.begin_fill()
pen.circle(30)
pen.end_fill()

pen.right(90)
pen.forward(30)
pen.up()
pen.goto(60,210)
pen.down()
pen.left(180)
pen.forward(30)

pen.up()
pen.goto(90,180)
pen.down()
pen.right(90)
pen.forward(30)

pen.up()
pen.goto(30,180)
pen.down()
pen.left(180)
pen.forward(30)

import math
pen.up()
pen.goto(60-(30*math.cos(45)),180+(30*math.sin(45)))
pen.down()
pen.right(45)
pen.forward(20)

pen.up()
pen.goto(60-(30*math.cos(45)),180-(30*math.sin(45)))
pen.down()
pen.left(90)
pen.forward(20)

pen.up()
pen.goto(60+(30*math.cos(45)),180-(30*math.sin(45)))
pen.down()
pen.left(90)
pen.forward(20)

pen.up()
pen.goto(60+(30*math.cos(45)),180+(30*math.sin(45)))
pen.down()
pen.left(90)
pen.forward(20)

# Vẽ ngôi nhà
pen.up()
pen.goto(-300,-300)
pen.down()
pen.right(45)
pen.fillcolor('blue')
pen.begin_fill()
pen.forward(200)
pen.left(90)
pen.forward(250)
pen.left(90)
pen.forward(200)
pen.left(90)
pen.forward(250)
pen.left(90)
pen.end_fill()
# Vẽ cửa
pen.up()
pen.goto(-240,-300)
pen.down()
pen.fillcolor('green')
pen.begin_fill()
pen.forward(80)
pen.left(90)
pen.forward(100)
pen.left(90)
pen.forward(80)
pen.left(90)
pen.forward(100)
pen.left(90)
pen.end_fill()
# Vẽ mái nhà
pen.up()
pen.goto(-300,-50)
pen.down()
pen.fillcolor('pink')
pen.begin_fill()
pen.forward(200)
pen.left(140)
pen.forward(100/math.cos(40))
pen.left(40)
pen.goto(-300,-50)
pen.end_fill()


turtle.done()
```
