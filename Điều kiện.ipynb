## Câu lệnh if

Nhập vào số điểm cuối kỳ của học sinh, nếu số điểm > 8.5 in ra "Bạn được tuyển thẳng"



```python
mark = float(input("Nhập điểm cuối kỳ: "))
if mark >= 8.5:
    print("Bạn được tuyển thẳng!")
```

### Thực hành:
Viết chương trình nhập vào môn học hiện tại. Nếu môn học là python thì in ra "I'm studying python"

```python
subject = input("Enter your class: ")
if subject == "python":
    print("I'm studying python")
```

## Câu lệnh if..else

```python
mark = float(input("Nhập điểm cuối kỳ: "))
if mark >= 8.5:
    print("Bạn được tuyển thẳng!")
else:
    print("Bạn chưa đủ điểm")
```

### Thực hành:
Viết chương trình nhập vào một số nguyên và cho biết số nguyên đó là số chẵn hay số lẻ.

```python
number = int(input("Please input a number: "))
if number % 2 == 1:
    print(f"{number} is an odd number")
else:
    print(f"{number} is an even number")
```

### Thực hành:
Viết chương trình nhập vào một chuỗi và cho biết trong chuỗi đó có chứa cụm từ "python" hay không.

```python
content = input("Please type here: ")
if "python" in content:
    print(f"'{content}' Includes python")
else:
    print(f"'{content}' does Not include python")
```

## Câu lệnh if lồng nhau

```python
# Nhập vào tổng số giờ làm và level của nhân viên, nhân viên sẽ được khen thưởng nếu như tổng số giờ làm > 800 giờ
# Mức độ khen thưởng như sau:
# Nếu là nhân viên level 1 thưởng 1000
# Nếu là nhân viên level 2 thưởng 2000

total_hour = int(input("Total working hour: "))
level = int(input("Level of employee: "))

if total_hour > 800:
    if level == 1:
        print("Bạn được nhận thưởng 1000 USD")
    else:
        print("Bạn được nhận thưởng 2000 USD")
else:
    print("Bạn chưa đạt tiêu chí nhận thưởng")
    
```

### Thực hành:
Viết chương trình nhập vào kết quả 2 lần rút thăm may mắn. Nếu cả 2 lần đều không trúng thì in ra kết quả "Chúc bạn may mắn lần sau". Nếu cả 2 lần đều trúng thì in ra "Bingo!!!"

## Câu lệnh if…elif…else

```python
# Viết chương trình nhập vào số ngày trong tuần (2,3,4,5,6,7,8) và in ra từ tiếng anh tương ứng với ngày đó

weekday = int(input("Today is: "))
if weekday == 2:
    print("Monday")
elif weekday == 3:
    print("Tuesday")
elif weekday == 4:
    print("Wednesday")
elif weekday == 5:
    print("Thursday")
elif weekday == 6:
    print("Friday")
elif weekday == 7:
    print("Saturday")
else:
    print("Sunday")
```

### Thực hành: 
Viết chương trình nhập vào số tháng trong năm và cho biết tháng đó có bao nhiêu ngày

```python
month = int(input("Nhập tháng trong năm: "))
if month in (1, 3, 5, 7, 8, 10, 12):
    print(f"Tháng {month} có 31 ngày")
elif month == 2:
    print(f"Tháng {month} có 28 ngày")
else:
    print(f"Tháng {month} có 30 ngày")
```
