```python
from tkinter import *
from tkinter import ttk
from tkinter.ttk import *
from tkinter import messagebox
from tkcalendar import Calendar, DateEntry
import csv 

bill = Tk ()
bill.title ('Construction Material Management System')
bill.geometry("850x400")
list_material = []  # lưu trữ danh sách các nguyên vật liệu xây dựng trong kho

f = open('Log.csv', 'w')
writer = csv.writer(f)

# tạo lớp
class Material:
    def __init__(self, code, name, quantity, date, imex):
        self.code = code
        self.name = name
        self.quantity = quantity
        self.date = date
        self.imex = imex # import or export
        
# hàm 
def add_material():
    add_name = check.get()
    Material_Table.insert(parent = '', index = 'end', values = (code.get(), name.get(), int(quantity.get()), date.get(), imex.get()))
    exist = False # giả định item không tồn tại trong list
    for i in range(len(list_material)):
        if list_material[i].name == add_name:
            exist = True
            if  imex.get() == 'Import': # nếu import, stock = quantity + nhập mới
                list_material.append(Material(code.get(), name.get(), int(quantity.get()), date.get(), imex.get()))
                list_material[i].quantity += int(quantity.get())
                messagebox.showinfo("Material", "Material name " + name.get() + " with Quantity of " + quantity.get() + " is added sucessfully")
            elif imex.get() == 'Export': # nếu import, stock = quantity + nhập mới
                list_material.append(Material(code.get(), name.get(), int(quantity.get()), date.get(), imex.get()))
                list_material[i].quantity -= int(quantity.get())
                messagebox.showinfo("Material", "Material name " + name.get() + " with Quantity of " + quantity.get() + " is removed sucessfully")
            break
    if exist == False:
        list_material.append(Material(code.get(), name.get(), int(quantity.get()), date.get(), imex.get()))
        messagebox.showinfo("Material", "Material name " + name.get() + " with Quantity of " + quantity.get() + " is added sucessfully")
    print([item.name for item in list_material])
    
    
def search_material():
    search_name = check.get()
    exist = False  # giả định là item không tồn tại trong list
    for item in list_material:
        if item.name == search_name:  # đã có kết quả tìm kiếm
            exist = True  # item đang tìm kiếm có tồn tại trong list
            messagebox.showinfo("Material", f"Quantity of {item.name} is {item.quantity}")
            break
    if exist == False:
        messagebox.showinfo("Material", f"{search_name} does not exist in list Material")    
        
def delete_material():
    delete_name = check.get()
    exist = False
    for i in range(len(list_material)):
        if list_material[i].name == delete_name:
            exist = True
            del list_material[i]
            messagebox.showinfo("Material", f"{delete_name} is deleted successfully")
            break
    if exist == False:
        messagebox.showinfo("Material", f"{delete_name} does not exist in list Material")
        
def exportlog():  
    with open("Log.csv", 'w', newline='') as f:
        header = ['ID', 'Material Name', 'Quantity', 'Date', 'Import/ Export']
        writer = csv.writer(f, dialect='excel')
        for row in writer:
            print(row)
            Material_Table.insert("", 'end', values = row)
        for row in writer:
            print(row)
            writer.writerows(row)
            f.close()
    messagebox.showinfo(f"Log of Import/ Export Material was downloaded!")
        
# create label
title_lbl = Label(bill, text = 'Construction Material Management System', font = ('Roboto Mono', 20, 'bold'))
manage_lbl = Label(bill, text = 'Managing Materials', font = ('Arial', 14, 'bold'), foreground = '#000099' )
id_lbl = Label(bill, text = 'Material ID:', font = ('Arial', 13, 'bold'))
name_lbl = Label(bill, text = 'Material name:', font = ('Arial', 13, 'bold'))
quantity_lbl = Label(bill, text = 'Quantity:', font = ('Arial', 13, 'bold'))
date_lbl = Label(bill, text = 'Date:', font = ('Arial', 13, 'bold'))
imex_lbl = Label(bill, text = 'Import/ Export:', font = ('Arial', 13, 'bold'))
searching_lbl = Label(bill, text = 'You can search for available stock by Material name here!', font = ('Arial', 13, 'italic'), foreground = '#990000')
matname_lbl = Label(bill, text = 'Material name:', font = ('Arial', 13, 'bold'))
log_lbl = Label(bill, text = 'Log of Material Import/ Export', font = ('Arial', 14, 'bold'), foreground = '#000099')

# create entry
code = Entry(bill)
name = Entry(bill)
quantity = Entry(bill)
date = Entry(bill)
check = Entry(bill)

# create combobox
imex = Combobox(bill, width = 10)
imex['value'] = ('Import','Export')
imex.current(0)

# create select date box
date = DateEntry(bill, width = 16, background = 'grey', foreground = 'black', borderwidth = 2, year = 2022)    

# create button
add_btn = Button(bill, text = 'Add', command = add_material)
delete_btn = Button(bill, text = 'Delete', command = delete_material)
search_btn = Button(bill, text = 'Search', command = search_material)
export_btn = Button(bill, text = 'Export Log',  command = exportlog)

# create table
Table_Frame = Frame(bill)
Table_Frame.place(x = 390, y = 80, width = 450, height = 280)
        
scroll = Scrollbar(Table_Frame, orient=VERTICAL)
Material_Table = ttk.Treeview(Table_Frame, columns=("ID", "Material Name", "Quantity", "Date", "Import/ Export"), yscrollcommand = scroll.set)
scroll.pack(side = RIGHT, fill = Y)
scroll.config(command = Material_Table.yview)
        
Material_Table.heading("ID", text = "ID")
Material_Table.heading("Material Name", text = "Material Name")
Material_Table.heading("Quantity", text = "Quantity")
Material_Table.heading("Date", text = "Date")
Material_Table.heading("Import/ Export", text = "Import/ Export")
        
Material_Table['show'] = 'headings'
Material_Table.column("ID", width = 40)
Material_Table.column("Material Name", width = 100)
Material_Table.column("Quantity", width = 100)
Material_Table.column("Date", width = 100)
Material_Table.column("Import/ Export", width = 100)

Material_Table.pack(fill=BOTH, expand=1)

# place
title_lbl.place(x = 260, y = 10)
id_lbl.place(x = 30, y = 86)
code.place(x = 150, y = 80)
name_lbl.place(x = 30, y = 116)
name.place(x = 150, y = 110)
quantity_lbl.place(x = 30, y = 146)
quantity.place(x = 150, y = 140)
date_lbl.place(x = 30, y = 176)
date.place(x = 150, y = 170)
imex_lbl.place(x = 30, y = 206)
imex.place(x = 150, y = 200)
add_btn.place(x = 150, y = 230)
delete_btn.place(x = 250, y = 230)
searching_lbl.place(x = 20, y = 270)
matname_lbl.place(x = 30, y = 306)
check.place(x = 150, y = 300)
search_btn.place(x = 250, y = 330)

manage_lbl.place(x = 30, y = 50)
log_lbl.place(x = 390, y = 50)
export_btn.place(x = 733, y = 50)

bill.mainloop ()
```
