```python
a = float(input("Average grade of a: "))
b = float(input("Average grade of b: "))
c = float(input("Average grade of c: "))

if a > b and a > c:
    print(f"A has the highest average grade: {a}")
elif b > c:
    print(f"B has the highest average grade: {b}")
else:
    print(f"C has the highest average grade: {c}")
```

## Cấu trúc lặp while

### Ví dụ: In ra tất cả số chẵn từ 0 đến n

```python
n = int(input("Input n: "))
count = 0
while count <= n:
    print(count)
    count += 2
```

### Ví dụ: In ra tất cả số lẻ từ n về 1

```python
n = int(input("Input n: "))
while n >= 1:
    if n % 2 == 1:
        print(n)
    n -= 1
```

### Ví dụ: Đảo ngược chuỗi

```python
# Input abcdef
# Output fedcba
```

```python
string = input("Input string: ")
index = len(string) - 1

while index >= 0:
    print(string[index], end="")
    index -= 1
```

### Lệnh Break

Cú pháp:

`break`

Lệnh break sẽ giúp hủy vòng lặp đang thực hiện

##### Lưu ý:

##### - Lệnh break <b>BẮT BUỘC</b> phải được đặt trong vòng lặp
##### - Lệnh break sẽ hủy vòng lặp gần nhất (trong trường hợp có nhiều vòng lắp lồng vào nhau)

#### Ví dụ

Sử dụng lệnh `while (True)` và break và in ra 10 lần câu "Hello mọi người"

```python
while True:
    print("Hello World")
    break
```

## Lệnh continue

```python
# In ra những số chia hết cho 4 trong khoảng từ 0 đến 20, từ số 8
count = 0
while count <= 20:
    count += 4
    if count == 8
        continue
    print(count)
    
```
